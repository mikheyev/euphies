# Esimating genome size using kmers
## General background

  - The earliest implementation of the idea below seems to be in the BGI giant panda paper, but I would like to know whether this is true
  - The basic idea here is that genome coverage (**C**) is the ratio between the amount of sequence generated (**S**) and genome size (**G**):

$$ C = S/G $$

  - We know how much sequence we have, but coverage requires a bit of trickery to estimate. Kmer coverage (**Ck**) for every kmer (**k**) can be computed as the product of nucleotide coverage (**C**) that is scaled by the probability of such a kmer lying within a read of a given length (**L**):
  
$$ Ck = C * (L - k + 1 ) / L $$

  - **k** and **L** are given, and if we can find **Ck**, can solve both equations and get a genome size estimate. Fortunately, kmer coverage can be estimated by looking a the frequency spectrum of all kmers found within the raw reads, as follows.

### Estimating genome size for the little fire ant (*Wasmannia auropunctata*)
I already know that this genome is about 314MB via flow cytometry, so let's see whether we can estimate it using a data set of 12,950,066,800 nucleotides sequenced using 50 bp reads.

First, I used [jellyfish](http://www.cbcb.umd.edu/software/jellyfish/) to compute kmer frequency histograms. This analysis picks up after that is done. The jellyfish script can be found [here](https://github.com/mikheyev/euphies/blob/master/joint/jellyfish.sh). It was designed to run on a large-memory server.

```python
import brewer2mpl
import gzip, glob,re
from collections import defaultdict
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pdb
%pylab inline
```

Read text results from *jellyfish histo* into an array. The text files have a column for frequency and another for the count.

```python
#read table dimensions
def get_counts(path):
    """
    This function takes a path to gzipped files containing the frequncy histograms. 
    It assumes that they have a kmer number somewhere in the file name, and two columns: coverage and count
    The function returns a list of kmers and a numpy array with their frequency spectra
    """
    kmer_counts = defaultdict(list)
    kmers = []
    maxlen = 0
    files = glob.glob(path)
    for file in files:
        kmers.append(int(re.search('[0-9]+',file).group(0)))
        infile = gzip.open(file)
        last_bin = int(infile.readlines()[-1].split()[0])
        if maxlen < last_bin:
            maxlen = last_bin
        infile.close()
    counts = np.zeros((maxlen,len(files)), dtype=np.int)
    
    for idx_file,file in enumerate(files):
        infile = gzip.open(file)
        for idx_bin,line in enumerate(infile):
            (bin, count) = map(int,line.rstrip().split())
            counts[idx_bin,idx_file] = count
        infile.close()
    return kmers,counts

was_kmer, was_counts = get_counts("data/assembly/was_*gz")
```

#### Plot kmer coverages

```python
#select less ugly set of colors
bmap = brewer2mpl.get_map('Set2', 'qualitative', len(was_kmer))
colors = bmap.mpl_colors
matplotlib.rcParams['axes.color_cycle'] = colors

#plot the kmer coverage distributions
maxrange = 50
cutoff = 5 # coverage cutoff for kmers

for i in range(3):
    plt.semilogy(range(0,maxrange), was_counts[range(0,maxrange),i], linestyle='-',color=colors[i])
legend(was_kmer)

#add markers 
#Note: I don't want to do this loop twice, but otherwise the legend gets messed up, and I can't be bothered to figure out why
kmer_coverages = {}
for i in range(3):  
    peak = max(was_counts[range(cutoff,maxrange),i])
    coverage = np.where(was_counts[range(0,maxrange),i] == peak)[0]
    kmer_coverages[was_kmer[i]] = int(coverage)
    plt.semilogy(np.where(was_counts[range(0,maxrange),i] == peak)[0],peak,"|",color="black", markersize=20)
    print "kmer %i: kmer covereage %i" % (was_kmer[i], kmer_coverages[was_kmer[i]])
grid(True)
plt.xlabel("kmer coverage")
plt.ylabel("kmer count")
plt.show()
```

This distribution should have a peak corresponding to the mean kmer coverage. In reality, we always also have a left sholder corresponding to error kmers, **i.e.**, kmers 

Larger kmers have lower coverage, whcih makes sense.

We're now ready to estimate genome coverage by solving the kmer coverage equation for **C**
$$ C = Ck * L / = (L - k + 1 ) $$

```python
read_len = 50.0
genome_coverages = {}
for kmer in kmer_coverages:
    genome_coverages[kmer] = kmer_coverages[kmer] * read_len / (read_len - kmer + 1)
    print "kmer %i: genome coverage %i" % (kmer, genome_coverages[kmer])
```

With these data in hand, we can make genome size estimations

```python
import locale
locale.setlocale(locale.LC_ALL, 'en_US')
data_size = 12950066800.0  #amount of sequenced reads
for g in  genome_coverages:
    print "kmer %i: %s bp" % (g, locale.format("%d", data_size/genome_coverages[g], grouping=True))
```

#### Take home notes
It looks like the analysis is sensitive to kmer size and coverage, with the smallest kmer producing the best results. This is likely due to how well the kmer coverage is estimated. For instance, for large kmers and small coverage values, you can't estimate the kmer coverage well, introducing error into this computation.

I also asked jellyfish to do some filtering, which may have resulted in some bases being thrown out, which would lead the genome size to be over-estimated.

## Using the same approach to estimate the genome sizes of two butterflies with unknown genomes

### Looking at assembly quality
This is based at ABYSS assemblies across a range of kmers from 71 to 95. We can't extend the max kmers, since the Illumina on the HiSeq 2000 top out at 100. 

```python
def read_stats_scaffolds(path):
    #Take a path to gzipped stats files, and get a dictionary of stats.
    stats = {}
    files = glob.glob(path)
    for file in files:
        kmer = int(re.search('[0-9]+',file).group(0))
        infile = gzip.open(file)
        text = infile.readlines()
        infile.close()
        scaffolds = text[-1].split()
        stats[kmer] = [int(scaffolds[5]), float(scaffolds[8])]
    return stats

aurinia_stats = read_stats_scaffolds("data/assembly/aurinia-stats*gz")
editha_stats = read_stats_scaffolds("data/assembly/editha-stats*gz")
```

```python
au = plt.plot(aurinia_stats.keys(),map(lambda x: aurinia_stats[x][0], aurinia_stats.keys()), "r-")
ed = plt.plot(editha_stats.keys(),map(lambda x: editha_stats[x][0], editha_stats.keys()), "b-")
plt.legend(["aurinia","editha"],'upper left')
plt.title("N50 size")
plt.xlabel("kmer")
plt.ylabel("bp")
plt.show()
```

```python
au = plt.plot(aurinia_stats.keys(),map(lambda x: aurinia_stats[x][1], aurinia_stats.keys()), "r-")
ed = plt.plot(editha_stats.keys(),map(lambda x: editha_stats[x][1], editha_stats.keys()), "b-")
plt.legend(["aurinia","editha"],'upper left')
plt.title("assembly size")
plt.xlabel("kmer")
plt.ylabel("bp")
plt.show()
```

It looks like we're asymptotically approaching an assembly size of something over 1G, which is in the ballpark of the predicted genome sizes of about 1.5G.
